#Step 2. Pull in eviction data from NYC OpenData API

**Note eviction data is filtered to just Neighborhood Tabulation Areas (NTAs) of Bushwick North and Bushwick South**

```python
#BUSHWICK NORTH

#'None' in place of application token, and no username or password bc public dataset
client = Socrata("data.cityofnewyork.us", None)

#filter to just "Bushwick North"
results = client.get("6z8x-wfk4", nta="Bushwick North")

#convert to pandas DataFrame
evictions_bushwick_north = pd.DataFrame.from_records(results)
```

```python
#BUSHWICK SOUTH

#'None' in place of application token, and no username or password bc public dataset
client = Socrata("data.cityofnewyork.us", None)

#filter to just "Bushwick South"
results = client.get("6z8x-wfk4", nta="Bushwick South")

#convert to pandas DataFrame
evictions_bushwick_south = pd.DataFrame.from_records(results)
```

```python
evictions_bushwick_south.head()
```

```python
#combine Bushwick North and Bushwick South dataframes into one Bushwick dataframe
all_bushwick_data = [evictions_bushwick_north,evictions_bushwick_south]
evictions_bushwick = pd.concat(all_bushwick_data)
evictions_bushwick
```

```python
#check for missing values

def missing_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
    '% of Total Values', ascending=False).round(1)
    print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
            " columns that have missing values.")
    return mis_val_table_ren_columns

missing_values_table(evictions_bushwick)
```

#Step 3. Filter to just residential evictions from 2017-2023

**I'm dropping observations from 2024, because I want only full years' worth of data for my map**

```python
#keep only residential evictions
evictions_bushwick = evictions_bushwick[evictions_bushwick['residential_commercial_ind'] == 'Residential']
```

```python
#drop if the eviction is still pending (i.e., there's no date for when the eviction was executed)
#there shouldn't be any observations that meet this criteria, since I had no values for this column when I ran missing_values_table
evictions_bushwick.dropna(subset=['executed_date'])
```

```python
#pull out year from executed_date column
from datetime import datetime

def extract_year(date_str):
    return pd.to_datetime(date_str).year

evictions_bushwick['Year'] = evictions_bushwick['executed_date'].apply(extract_year)
```

```python
evictions_bushwick['Year'] = evictions_bushwick['Year'].astype(int)
```

```python
#now drop observations from 2024
evictions_bushwick = evictions_bushwick.drop(evictions_bushwick[evictions_bushwick.Year == 2024].index)
```

```python
#sort descending by date to make sure all 2024 observations were dropped
evictions_bushwick.sort_values(by=['executed_date'],ascending=False,inplace=True)
evictions_bushwick.head()
```

#Step 4. Delete duplicates

**Tenants can get a stay of eviction even after the Marshall's Notice is issued.**

If records have the same value for court_index_number (i.e., the same plaintiff is being sued in Housing Court to be evicted from the same apartment), I will keep only the most recent record. That way, an eviction that was stayed and then executed won't show up as two separate evictions.


```python
#find dupes for court_index_number and keep first (I already descending sorted by date, so the most recent records will be first.)
evictions_bushwick.drop_duplicates(subset='court_index_number', keep='first', inplace=True)
```

#Step 5. Clean data to use in map and export

**Create one column for address.**

I tried cleaning up the apartment number column, but it's super inconsistent. (Some values are 1ST FLOOR versus 2L verus 211, so it's hard to get the capitalization right.)

Also, sometimes apartment number is in the address line too. If I try to concetenate the address, it will turn into something like "325 Knickerbocker Ave Apt 2L, 2L, Brooklyn, NY 11237"

**I'm going to drop apartment number from the Address column, so I can focus on making my map (and not on trying to clean this column).**

```python
#change capitalization sytle to just first letter capitalized for street address
evictions_bushwick['Address'] = evictions_bushwick['eviction_address'].str.title()
evictions_bushwick.head()
```

```python
#create one column for full address -- without apartment number :(
evictions_bushwick['Full-Address'] = evictions_bushwick['Address'] + ', Brooklyn, NY ' + evictions_bushwick['eviction_zip']
evictions_bushwick.head()
```

```python
#convert 'date' column to simple datetime type
evictions_bushwick['executed_date'] = pd.to_datetime(evictions_bushwick['executed_date'])

#create a new column with formatted dates (month written out, day, year)
evictions_bushwick['Date'] = evictions_bushwick['executed_date'].dt.strftime('%B %d, %Y')
evictions_bushwick.head()
```

```python
#add indicator for if an eviction occurred during the COVID-19 Moratorium
evictions_bushwick['During-Moratorium'] = (
    (evictions_bushwick['executed_date'] >= '2020-03-07') & (evictions_bushwick['executed_date'] <= '2022-01-15')).astype(int)
evictions_bushwick.head()
```

```python
evictions_bushwick = evictions_bushwick.rename(
    {'longitude': 'Longitude',
     'latitude':'Latitude'}, axis="columns"
)
```

```python
#keep fields that I will need for map -- Address, Date, Year, Longitude, Latitude
evictions_bushwick = evictions_bushwick[['Date','Address','Year','Longitude','Latitude']]
evictions_bushwick.head()
```

```python
#download geojson for map
gdf = gpd.GeoDataFrame(evictions_bushwick, geometry=gpd.points_from_xy(evictions_bushwick['Longitude'], evictions_bushwick['Latitude']))
gdf.to_file('evictions_bushwick_cleaned.geojson', driver='GeoJSON')
files.download('evictions_bushwick_cleaned.geojson')
```
